```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('Lenna.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

size = 4
kernel = np.ones((size,size), np.float32)/(size **2)

print(kernel)

dst = cv2.filter2D(image, -1, kernel)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image = cv2.imread('Lenna.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

mask1 = np.array([[-1,0,0], [0,0,0], [0,0,1]])
mask2 = np.array([[1,1,1], [1, -8, 1], [1,1,1]])

out1 = cv2.filter2D(gray, -1, mask1)
out2 = cv2.filter2D(gray, -1, mask2)

cv2_imshow(gray)
cv2_imshow(out1)
cv2_imshow(out2)

cv2.waitKey(0)
```

```python
image = cv2.imread('Lenna.jpg')
rows, cols = image.shape[0:2]

d45 = 45.0 + np.pi / 180
d90 = 90.0 + np.pi / 180

n45 = np.float32([[np.cos(d45), -1*np.sin(d45), rows//2],
                  [np.sin(d45), np.cos(d45), -1*cols//4]])
n90 = np.float32([[np.cos(d90), -1*np.sin(d90), rows],
                  [np.sin(d90), np.cos(d90), 0]])

r45 = cv2.warpAffine(image, n45, (cols, rows))
r90 = cv2.warpAffine(image, n90, (cols, rows))

cv2_imshow(image)
cv2_imshow(r45)
cv2_imshow(r90)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
n45 = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 0.5)
n90 = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1.5)

image45 = cv2.warpAffine(image, n45,(cols, rows))
image90 = cv2.warpAffine(image, n90,(cols, rows))

cv2_imshow(image)
cv2_imshow(image45)
cv2_imshow(image90)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('Lenna.jpg')
height, width = image.shape[:2]

m_small = np.float32([[0.5, 0, 0],
                      [0,0.5,0]])
m_big = np.float32([[2, 0, 0],
                      [0,2,0]])

dist1= cv2.warpAffine(image, m_small, (int(height*0.5), int(width*0.5)))
dist2= cv2.warpAffine(image, m_big, (int(height*2), int(width*2)))

dist3 = cv2.warpAffine(image, m_small, (int(height*0.5), int(width*0.5)), None, cv2.INTER_AREA)
dist4 = cv2.warpAffine(image, m_big, (int(height*2), int(width*2)), None, cv2.INTER_CUBIC)

cv2_imshow(image)
cv2_imshow(dist1)
cv2_imshow(dist2)
cv2_imshow(dist3)
cv2_imshow(dist4)
```
